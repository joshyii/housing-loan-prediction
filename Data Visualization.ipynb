# **Visualization of combination of attributes**

### "Loan_Status" and "Gender"

```python
print(pd.crosstab(loan["Gender"],loan["Loan_Status"]))
Gender = pd.crosstab(loan["Gender"],loan["Loan_Status"])
Gender.div(Gender.sum(1).astype(float),axis=0).plot(kind="bar",stacked=True,figsize=(4,4))
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.title("Stacked Bar Chart - Gender & Loan Status")
```

### Loan_Status and ApplicantIncome 

```python
loan.groupby("Loan_Status")['ApplicantIncome'].mean().plot.bar()
```

```python
quartile = np.array([0])
quartile = np.append(quartile,loan["ApplicantIncome"].quantile([0.25,0.5,0.75]))
group=['Low','Average','High']
loan['Income_bin']=pd.cut(loan['ApplicantIncome'], quartile,labels=group)

print(pd.crosstab(loan["Income_bin"],loan["Loan_Status"]))
Income_bin = pd.crosstab(loan["Income_bin"],loan["Loan_Status"])
Income_bin.div(Income_bin.sum(1).astype(float),axis=0).plot(kind="bar",stacked=True,figsize=(4,4))
plt.xlabel("ApplicantIncome")
plt.ylabel("Percentage")
plt.title("Stacked Bar Chart - Applicant Income & Loan Status")
```

### "Loan_Status" and "Total Income"

```python
loan["TotalIncome"]=loan["ApplicantIncome"]+loan["CoapplicantIncome"]

quartile = np.array([0])
quartile = np.append(quartile,loan["TotalIncome"].quantile([0.25,0.5,0.75]))

group=['Low','Average','High']
loan["TotalIncome_bin"]=pd.cut(loan["TotalIncome"], quartile ,labels=group)
print(pd.crosstab(loan["TotalIncome_bin"],loan["Loan_Status"]))
TotalIncome = pd.crosstab(loan["TotalIncome_bin"],loan["Loan_Status"])
TotalIncome.div(TotalIncome.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,figsize=(4,4))
plt.xlabel("TotalIncome")
plt.ylabel("Percentage")
plt.title("Stacked Bar Chart - Total Income & Loan Status")
```

### "Loan_Status" and "Loan Amount"

```python
quartile = np.array([0])
quartile = np.append(quartile,loan["LoanAmount"].quantile([0.25,0.5,0.75]))
loan["LoanAmount_bin"]=pd.cut(loan["LoanAmount"],quartile,labels=group)
print(pd.crosstab(loan["LoanAmount_bin"],loan["Loan_Status"]))
LoanAmount=pd.crosstab(loan["LoanAmount_bin"],loan["Loan_Status"])
LoanAmount.div(LoanAmount.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,figsize=(4,4))
plt.xlabel("LoanAmount")
plt.ylabel("Percentage")
plt.title("Stacked Bar Chart - Loan Amount & Loan Status")
```
